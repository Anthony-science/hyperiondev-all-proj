```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
df.head()
```

* Compare the average income based on ethnicity.

```python
print(df.groupby('Ethnicity')['Income'].mean())
```

* On average, do married or single people have a higher balance?

```python
marital_income = df.groupby('Married')['Income'].mean()
df['Married'] = df['Married'].replace({'Yes': 'Married', 'No': 'Not Married'})
max_married_status = marital_income.idxmax()
print(f"The highest income group is: {max_married_status}. Average income for this group is:", marital_income.max().round(2)) 
```

* What is the highest income in our dataset?

```python
print(f"The highest income is: ", df['Income'].max().round(2))
```

* What is the lowest income in our dataset?

```python
print(f"The lowest income is: ", df['Income'].min().round(2))
```

* How many cards do we have recorded in our dataset? (Hint: use `sum()`)

```python
print(f"Total number of cards in the dataset: {df['Cards'].sum()}")
```
